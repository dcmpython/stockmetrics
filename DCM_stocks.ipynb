```python
#A simple program to teach people the basics of P/E multiples, the cost of equity, and to spot opportunities in the market
#Thanks to Dr Barone, Dr Xu, and Dr Olsen for their inspiration in helping me develop this program

# This text sets the stage for the user to understsand P/E multiples, cost of equity, and simple financial terms for the business
print("In Dr. Barone's class we learned all about P/E multiples, which are a simple yet effective way to look at the relative valuation of stocks!")
print()
print("The simple formula for a P/E ratio is 1 divided by: (r minus g) where:")
print("  =>r is the firm's cost of capital and")
print("  =>g is the expected growth rate")
print()
print("In Dr. Xu's class we learned all about forecasting a firm's cost of equity:")
print("A firm's cost of equity (r) is the risk free-rate * Beta times (expected market return above the risk free rate)")
print()
print("In Dr. Olsen's Python class, we learned how to create lists, tuples, and dictonaries - which are similar to small databaes - and basic mathematical calculations using Python for data analysis.")
print("Since python is being integrated in Excel and we have ChatGPT to help us code, we can now all become superstar Financial Analysts")
print()
print("To get started, which stock would you like to look at first. We have four stocks we can analyze:")
print("1-$CRM")
print("2-$MSFT")
print("3-$AAPL")
print("4-$GOOGL")
print()
print("NOTE: According to ChatGPT, dictionaries allow you to associate keys (stock tickers) with values (company metrics).")
print("This key-value structure makes it easy to retrieve and update the metrics for a specific stock ticker")

#Below is the code for our starting database
#For this simple database we have decided to use a dictionary
# ChatGPT explains why below
#Key-Value Structure: Dictionaries allow you to associate keys (stock tickers) with values (company metrics).
#This key-value structure makes it easy to retrieve and update the metrics for a specific stock ticker

import math
import statistics

stock_metrics = {
    'CRM': {
        'price': 202.73,
        'EPS_2022': 5.24,
        # You can add more metrics for MSFT here
    },
   'MSFT': {
        'price': 312.79,
        'EPS_2022': 9.68,
        # You can add more metrics for MSFT here
    },
     'AAPL': {
        'price': 170.43,
        'EPS_2022': 6.11,
        # You can add more metrics for MSFT here
    },
     'GOOGL': {
        'price': 130.54,
        'EPS_2022': 4.56,
        # You can add more metrics for MSFT here
    },
}

# Calculate the P/E ratio for each ticker
for ticker, metrics in stock_metrics.items():
    if 'EPS_2022' in metrics and metrics['EPS_2022'] != 0:
        pe_ratio = metrics['price'] / metrics['EPS_2022']  # Calculate P/E ratio based on price and EPS
        metrics['pe_ratio'] = pe_ratio  # Add P/E ratio to the dictionary

ticker = input("Please enter a ticker without the $ sign here: ")
if ticker == 'CRM':
  print()
  print(f'You picked {ticker}. They are a great CRM tool used by many salespeople. Let us look at their P/E multiple.')
  if 'CRM' in stock_metrics and 'EPS_2022' in stock_metrics['CRM']:
      print(f'According to our Python dictionary, CRM EPS in 2022 was ${stock_metrics["CRM"]["EPS_2022"]}')
      #retrieve the EPS values and prices values and store them outside the dictionary as variables
      price = stock_metrics["CRM"]["price"]
      eps = stock_metrics["CRM"]["EPS_2022"]
      pe_ratio = price / eps
      print(f'With {ticker} recent price of ${price}, that equates to a P/E ratio of {pe_ratio:.1f}x on what Investors call a trailing twelve month basis, or TTM for short.')
      print()
  else:
      print("Sorry, we don't have price and/or EPS data for CRM.")
if ticker == 'MSFT':
  print()
  print(f'You picked {ticker}. The CEO has turned them into a powerhouse for corporate productivity. Let us look at their P/E multiple.')
  if 'MSFT' in stock_metrics and 'EPS_2022' in stock_metrics['MSFT']:
      print(f'According to our Python dictionary, MSFT EPS in 2022 was ${stock_metrics["MSFT"]["EPS_2022"]}')
      #retrieve the EPS values and prices values and store them outside the dictionary as variables
      price = stock_metrics["MSFT"]["price"]
      eps = stock_metrics["MSFT"]["EPS_2022"]
      pe_ratio = price / eps
      print(f'With {ticker} recent price of ${price}, that equates to a P/E ratio of {pe_ratio:.1f}x on what Investors call a trailing twelve month basis, or TTM for short.')
      print()
  else:
      print("Sorry, we don't have price and/or EPS data for MSFT.")
if ticker == 'AAPL':
  print()
  print(f'You picked {ticker}. Steve Jobs left us a good business with this one. Let us now look at their P/E multiple.')
  if 'AAPL' in stock_metrics and 'EPS_2022' in stock_metrics['AAPL']:
      print(f'According to our Python dictionary, AAPL EPS in 2022 was ${stock_metrics["AAPL"]["EPS_2022"]}')
      #retrieve the EPS values and prices values and store them outside the dictionary as variables
      price = stock_metrics["AAPL"]["price"]
      eps = stock_metrics["AAPL"]["EPS_2022"]
      pe_ratio = price / eps
      print(f'With {ticker} recent price of ${price}, that equates to a P/E ratio of {pe_ratio:.1f}x on what Investors call a trailing twelve month basis, or TTM for short.')
      print()
  else:
      print("Sorry, we don't have price and/or EPS data for AAPL.")
if ticker == 'GOOGL':
  print()
  print(f'You picked {ticker}. Will ChatGPT overtake Google Search. Let us look at their P/E multiple to see what the market thinks.')
  if 'GOOGL' in stock_metrics and 'EPS_2022' in stock_metrics['GOOGL']:
      print(f'According to our Python dictionary, GOOGL EPS in 2022 was ${stock_metrics["GOOGL"]["EPS_2022"]}')
      #retrieve the EPS values and prices values and store them outside the dictionary as variables
      price = stock_metrics["GOOGL"]["price"]
      eps = stock_metrics["GOOGL"]["EPS_2022"]
      pe_ratio = price / eps
      print(f'With {ticker} recent price of ${price}, that equates to a P/E ratio of {pe_ratio:.1f}x on what Investors call a trailing twelve month basis, or TTM for short.')
      print()
  else:
      print("Sorry, we don't have price and/or EPS data for GOOGL.")

print()
print("-----------------------------------------------------------------------------")

# Create a list to store P/E ratios for the companies
pe_ratios = []

# Loop through the dictionary and calculate the average and median P/E ratios
for ticker, metrics in stock_metrics.items():
    if 'pe_ratio' in metrics:
        pe_ratio = metrics['pe_ratio']
        pe_ratios.append(pe_ratio)

# Calculate and print average P/E ratio
average_pe_ratio = statistics.mean(pe_ratios)

# Calculate median P/E ratio
median_pe_ratio = statistics.median(pe_ratios)

# Print a quasi comp table to get the user used to seeing things like Wall Street does
print(f'You do NOT have to work on Wall Street to use comp tables')
print("\nCompany Comp Table")
print(f"{'Ticker':<15} {'TTM P/E Ratio':<15}")
for ticker, metrics in stock_metrics.items():
    if 'pe_ratio' in metrics:
        pe_ratio = metrics['pe_ratio']
        print(f"{ticker:<15} {pe_ratio:.1f}x")
    else:
        print(f"{ticker:<15} N/A")
print(f'---------------------')
print(f"Median for Group: {median_pe_ratio:.1f}x")
print(f"Average for Group: {average_pe_ratio:.1f}x")

#Need some now what prompt like a menu
print("--------------------------------------------------------------")
print("Finally, let's go back to the main menu to see what else we can do")
print("For now let's just see what is all in our expanded dictionary")
print("In other word's, which tickers do we have and which metric's do we have")
print("We will print this in an unstructured form using a nested for loop")
print()

# Print the updated "unstructured" dictionary with P/E ratios
for ticker, metrics in stock_metrics.items():
    print(f'Ticker: {ticker}')
    for metric, value in metrics.items():
        print(f'{metric}: {value:.2f}')
```
